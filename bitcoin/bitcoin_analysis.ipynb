```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Descargar datos histricos de Bitcoin
btc_data = yf.download('BTC-USD', start='2020-01-01', end='2023-12-31')

# Calcular retornos diarios
btc_data['Daily Return'] = btc_data['Adj Close'].pct_change()

# Calcular mtricas clave
mean_return = btc_data['Daily Return'].mean()
volatility = btc_data['Daily Return'].std()
sharpe_ratio = mean_return / volatility * np.sqrt(252)

# Visualizacin
plt.figure(figsize=(12, 6))
plt.plot(btc_data['Adj Close'], label='Bitcoin Price')
plt.title('Bitcoin Historical Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

print(f"Mean Daily Return: {mean_return:.5f}")
print(f"Volatility (Std Dev): {volatility:.5f}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```
